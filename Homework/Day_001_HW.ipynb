## 顯示單一顏色通道的圖片 

```python
import cv2
```

```python
b = img.copy()
# set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0

g = img.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = img.copy()
# set blue and green channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0


# Blue
cv2.imshow( 'Blue Channel', b )

# Green
cv2.imshow( 'Green Channel', g )

# Red
cv2.imshow( 'Red Channel', r )
cv2.waitKey( 10000 )
cv2.destroyAllWindows() 
```
